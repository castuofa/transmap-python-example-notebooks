# Using the TransMAP API

### Step 1: Import the TransMAP class from the transmap module.

```python
from transmap import TransMAP, Basemaps
```

### Step 2: Create a new instance of the TransMAP class.
The TransMAP class requires that the user supply the URL to the active transmap REST API, the current api is located at https://oak.cast.uark.edu/metadata-api

```python
tm = TransMAP('https://oak.cast.uark.edu/metadata-api')
```

### Step 3: Search for datasets using the search function.
Below I have searched for the text 'AIS' to find all the AIS related datasets currently registered in TransMAP

```python
datasets = tm.search('AIS')

for dataset in datasets:
    print(dataset.title)
```

### Step 4: Fetch the data. 
Here I have fetched the data for the AIS Freight dataset found via the search function above.
Specify a start and end data to choose data for a particular time period.

```python
data = tm.fetch(datasets[0], start_date= "8/29/2019 12:00:00 AM", end_date="8/30/2019 12:00:00 AM")
data.head()
```

<b>Note:</b> It is also possible to fetch data with only the dataset id.

```python
data = tm.fetch(dataset_id='627d23f7bf25367169a264ea', start_date= "8/29/2019 12:00:00 AM", end_date="8/29/2019 12:00:00 AM")
data.tail()
```

### Step 5: Inspect the dataset with a simple map view.
You can quickly view the data on a simple static map using .map()

```python
data.simple_map(size=(10,10), edgecolor="red")
```
